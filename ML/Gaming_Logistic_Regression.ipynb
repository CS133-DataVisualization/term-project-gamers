### TF-IDF Analysis on Game Descriptions
We will use the text in `detailed_description` to see if we can predict the `metacritic_score`.

```python
import re
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import accuracy_score, classification_report

# Preprocess Text
def preprocess_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s\-]", "", text)
    return text

df_rated['processed_description'] = df_rated['detailed_description'].apply(preprocess_text)

# Logistic Regression is a classification model, so turn metacritic scores into classes
def score_to_class(score):
    if score < 60:
        return "bad"
    elif score < 75:
        return "mixed"
    else:
        return "good"

df_rated['score_class'] = df_rated['metacritic_score'].apply(score_to_class)

#Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    df_rated['processed_description'],
    df_rated['score_class'],
    test_size=0.2,
    random_state=42,
    stratify=df_rated['score_class']
)

# Pipeline
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer(
        max_features=10000,
        stop_words='english',
        ngram_range=(1, 2),
        min_df=3
    )),
    ("logreg", LogisticRegressionCV(
        Cs=5,
        cv=5,
        max_iter=1000,
        n_jobs=-1,
        class_weight='balanced'
    ))
])


cv_scores = cross_val_score(
    pipeline,
    X_train,
    y_train,
    cv=5,
    scoring='accuracy'
)

print(f"\nCross-Validation Accuracy: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")

# Train Final Model
pipeline.fit(X_train, y_train)

# Evaluation Metrics
y_pred = pipeline.predict(X_test)

print("\nTest Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

### **Metrics Visualization**
Confusion Matrix to see specific misclassifications and bar chart to comparre precision, recall, and F1-score

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# 1. Confusion Matrix
cm = confusion_matrix(y_test, y_pred, labels=pipeline.classes_)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=pipeline.classes_,
            yticklabels=pipeline.classes_)
plt.title('Confusion Matrix: Predicted vs. Actual Scores')
plt.ylabel('Actual Category')
plt.xlabel('Predicted Category')
plt.show()

# 2. Metrics Bar Chart
report = classification_report(y_test, y_pred, output_dict=True)
metrics_df = pd.DataFrame(report).transpose().iloc[:3, :3] # Only get bad/good/mixed classes

metrics_df.plot(kind='bar', figsize=(10, 6))
plt.title('Model Metrics by Score Category')
plt.ylabel('Score (0.0 - 1.0)')
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Our model is decent at predicting high-rated games, but it seems that due to lack of data on poorly rated games it performs poorly on bad games. This is actually pretty interesting because this may suggest that critics have some positivity bias when it comes to reviews, and this insight supports our theory that critics aren't very harsh and games in general are simply rated fairly highly, it's rare that a game gets a very poor review or score.
